```python
def search(keyword):
    Entrez.email = 'adamsorbie@tum.de'
    handle = Entrez.esearch(db='pubmed', 
                            sort='relevance', 
                            retmax='100',
                            retmode='xml', 
                            term=keyword)
    results = Entrez.read(handle)
    return results
# search 
```

```python
def pubmed_fetch(id_list):
    ids = ','.join(id_list)
    Entrez.email = 'adamsorbie@tum.de'
    handle = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=ids)
    results = Entrez.read(handle)
    return results
```
